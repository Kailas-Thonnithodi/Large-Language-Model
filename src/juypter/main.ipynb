## Encoder and Decoder
* Encoding each token (text or letter) or character instance into a numeric value.
* Decoding is converting the mapped numeric value with original letter / token instance. 

```python
# Encoder
def encoder(chars: list[str]):
    string_to_integer = {string:index for index, string in enumerate(chars)}
    print(string_to_integer)
    return (lambda string_keys: [string_to_integer[string] for string in string_keys])

encoded = encoder(characters)
```

```python
# Decoder
def decoder(chars: list[int]):
    integer_to_string = {index:string for index, string in enumerate(chars)}
    print(integer_to_string)
    return (lambda index_keys: ''.join([integer_to_string[index] for index in index_keys]))

decoded = decoder(characters)
```
